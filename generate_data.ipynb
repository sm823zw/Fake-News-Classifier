```python
import re
import nltk
import pickle
import numpy as np
import pandas as pd
import tensorflow as tf
```

```python
physical_devices = tf.config.list_physical_devices('GPU')
tf.config.experimental.set_memory_growth(physical_devices[0], True)
```

```python
data = pd.read_csv('train.csv')
```

```python
data
```

```python
data = data.iloc[:, [1, 4]]
data.head()
```

```python
data = data.dropna()
data
```

```python
len(data)
```

```python
stemmer = nltk.stem.PorterStemmer()
stopwords = nltk.corpus.stopwords.words('english')
def clean_data(text):
    text = re.sub('[^a-zA-Z]',' ', text)
    text = text.lower()
    text = text.split()
    text = [stemmer.stem(j) for j in text if j not in stopwords]
    text = ' '.join(text)
    return text
```

```python
sentences = list(data['title'])
labels = list(data['label'])
```

```python
data = []
for i in sentences:
    text = clean_data(i)
    data.append(text)
```

```python
x = [data, labels]
```

```python
with open('title_train_data_oh.pickle', "wb") as file:
        pickle.dump(x, file)
```
